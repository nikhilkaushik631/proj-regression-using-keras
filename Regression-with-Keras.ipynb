Import Libraries

```python
import pandas as pd

from keras.models import Sequential
from keras.layers import Dense
```

Download dataset


```python
df = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/df.csv')
df.head()
```

```python
df.shape
```

Missing values


```python
df.describe()
```

```python
df.isnull().sum()
```

Splitting dataset


```python
columns = df.columns

X = df[columns[columns != 'Strength']] # all columns except Strength
Y = df['Strength'] # Strength column
```

```python
X.head()
```

```python
Y.head()
```

Normalization


```python
X_norm = (X - X.mean()) / X.std()
X_norm.head()
```

```python
n_cols = X_norm.shape[1] # number of predictors
```

#### Neural Network


```python

def regression_model():
   
    model = Sequential()
    model.add(Dense(50, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

Training and Testing


```python

model = regression_model()
```

```python
# fit the model
model.fit(X_norm, Y, validation_split=0.3, epochs=100, verbose=2)
```
